## 시군구별 이혼 사유

```python
divorce_pivot = pd.pivot_table(divorce_reason, index=['시도별', '항목'], values=divorce_reason.columns[2:], aggfunc='sum', margins=True)
divorce_pivot
```

```python
divorce_reason.columns[2:]
```

```python
divorce_melt = divorce_reason.melt(id_vars=['시도별', '항목'], var_name="발생 연도", value_name="발생 건수")
divorce_melt = divorce_melt.rename(columns={'시도별':'지역', '항목':'이혼 사유', '발생 연도':'연도', '발생 건수':'이혼 횟수'})
divorce_melt
```

```python
# 전국, 국외 데이터는 삭제하기
divorce_melt = divorce_melt.drop(index=divorce_melt.loc[divorce_melt['지역'] == '전국'].index)

# 이혼 사유의 계 데이터 삭제하기
divorce_melt = divorce_melt.drop(index=divorce_melt.loc[divorce_melt['이혼 사유']=='계'].index)

divorce_melt
```

```python
# int로 type 변경
divorce_melt['연도'] = divorce_melt['연도'].astype(int)

# - 값은 0으로 대체
divorce_melt['이혼 횟수'] = divorce_melt['이혼 횟수'].replace("-", 0)
divorce_melt['이혼 횟수'] = divorce_melt['이혼 횟수'].astype(int)

divorce_melt.info()
```

```python
divorce_melt.groupby('지역')['이혼 횟수'].sum().sort_values(ascending=False).plot(kind='bar', title='지역별 이혼 횟수', color='red')
```

```python
divorce_melt.groupby('이혼 사유')['이혼 횟수'].sum().sort_values(ascending=False).plot(kind='bar', title='이혼 사유별 이혼 횟수', color='green')
```

이혼 사유는 성격차이가 가장 많다는 결론을 얻었다.

## 이혼 사유별 지역별 이혼 횟수

```python
plt.figure(figsize=(16,4))
plt.title('이혼 사유별 횟수')
sns.countplot(data=divorce_melt, x='이혼 사유', hue='지역')
plt.legend(bbox_to_anchor=(1,1))
```

```python
plt.figure(figsize=(16,4))
sns.catplot(data=divorce_melt, x='이혼 사유', y='이혼 횟수', hue='지역')
plt.title('이혼 사유별 지역별 이혼 횟수')
```
