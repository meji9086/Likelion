# Preprocessing

```python
todate = ['shipping_limit_date', 'review_creation_date', 'order_purchase_timestamp', 'order_purchase_timestamp', 'order_delivered_carrier_date'   
            ,'order_delivered_customer_date', 'order_estimated_delivery_date', 'order_approved_at', 'review_answer_timestamp']

for d in todate:
    data[d] = pd.to_datetime(data[d])

data[todate].info()
```

```python
data = data.drop(['review_comment_title', 'review_comment_message'], axis=1)
data = data.dropna()
data.isnull().sum()
```

```python
dev_date = []

for d, a in zip(data['order_delivered_customer_date'], data['order_approved_at']):
    dev_date.append((d-a).days)

data['delivery_date'] = dev_date
data['delivery_date']
```

# Visualization

```python
review_count = data['review_score'].value_counts()
review_count = pd.DataFrame(review_count)
review_count.columns = ['count']
review_count = review_count.sort_index()
review_count
```

```python
plt.figure(figsize=(10,8))
ax=sns.barplot(data=review_count, x=review_count.index, y='count', palette='PuBu')
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in plt.gca().get_yticks()])
_=plt.title('전체 평점 점수별 리뷰 수')
_=ax.set(ylabel=None)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, height + 500, f'{int(height)}개', ha = 'center', size = 10)
```

```python
data['delivered_ymd'] = data['order_delivered_customer_date'].map(lambda x : x.strftime('%Y-%m-%d'))
data['estimated_ymd'] = data['order_estimated_delivery_date'].map(lambda x : x.strftime('%Y-%m-%d'))
```

```python
early = data[data['delivered_ymd'] < data['estimated_ymd']]
early = early['review_score'].value_counts()
early = pd.DataFrame(early)
early.columns = ['count']
early = early.sort_index()
early
```

```python
early.sum()
```

```python
plt.figure(figsize=(10,8))
ax=sns.barplot(data=early, x=early.index, y='count', palette='PuBu')
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in plt.gca().get_yticks()])
_=plt.title('예상일보다 빨리 도착했을 경우 평점 점수별 리뷰 수')
_=ax.set(ylabel=None)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, height + 500, f'{(int(height)/104501)*100:.0f}%', ha = 'center', size = 10)
```

```python
same = data[data['delivered_ymd'] == data['estimated_ymd']]
same = same['review_score'].value_counts()
same = pd.DataFrame(same)
same.columns = ['count']
same = same.sort_index()
same
```
