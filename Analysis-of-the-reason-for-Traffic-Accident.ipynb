# 데이터 전처리

```python
# 결측치 데이터는 0으로 대체
df = df.replace("-", 0)

# 필요없는 데이터 삭제하기
df = df.drop(df.index[[-1]])
df = df.drop(['합계','합계.1','합계.2'], axis=1)
df = df.drop(['법규위반_대'], axis=1)

# column 이름을 위해 변경하기
df['법규위반_소'][0] = '법규위반유형'

# 열 이름 바꾸기 
month = 0
for i in range(2, len(df.columns)):
    # 3개씩 끊기
    if (i % 3 == 2):
        month += 1
    # 10월달 이상은 앞에 0 제외    
    if (month >= 10):
        df.iloc[0][i] = f'{month}_{df.iloc[0][i]}'
    else:
        df.iloc[0][i] = f'0{month}_{df.iloc[0][i]}'

df = df.rename(columns=df.iloc[0])
df = df.drop(df.index[[0]])

# 타입 숫자->int 문자->category로 변경하기
for col in df.columns[:2]:
    df[col] = df[col].astype("category")

for col in df.columns[2:]:
    df[col] = pd.to_numeric(df[col], downcast="integer")


# 합계가 100보다 적은 데이터 및 필요없는 법규위반 삭제하기
df['합계'] = df.sum(axis=1)
df = df.drop(df[df['합계']<100].index)
df = df.drop(df[df['법규위반유형'].str.contains("기타")].index)
df = df.drop(df[df['법규위반유형'].str.contains("안전운전불이행")].index)
df = df.reset_index(drop=True)

# 법규위반유형_주야 합치기
df['법규위반유형'] = df['법규위반유형'].str.cat(df['주야'], sep='_')
df = df.drop('주야', axis=1)

# 합계 데이터 지우기
df = df.drop('합계', axis=1)

# Transpos 설정
df = df.T
df = df.rename(columns=df.iloc[0])
df = df.drop('법규위반유형')
df = df.reset_index()
df = df.rename(columns = {'index' : '유형'})

# 전처리한 데이터 출력
df
```

```python
# 시각화를 위한 데이터 만들기
df2 = df.melt(id_vars='유형', var_name='법규위반', value_name='사고수')

df2['법규위반유형'] = df2['법규위반'].map(lambda x : x.split("_")[0])
df2['주야'] = df2['법규위반'].map(lambda x : x.split("_")[1])
df2['월'] = df2['유형'].map(lambda x : x.split("_")[0])
df2['유형'] = df2['유형'].map(lambda x : x.split("_")[1])
df2 = df2.drop('법규위반', axis=1)

df2 = df2[['법규위반유형', '유형', '주야', '월', '사고수']]
df2
```
