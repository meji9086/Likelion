```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

def webtoon_rank(days=None):
    # 웹 페이지를 열고 소스코드를 읽어오는 작업
    url = "https://comic.naver.com/webtoon/weekday.nhn"
    res = requests.get(url)
    res.raise_for_status()

    soup = BeautifulSoup(res.text, "html.parser")

    day_lst = ['일', '월', '화', '수', '목', '금', '토']
    webtoon_rank = pd.DataFrame()

    # 요일 별 제목 가져오기 위한 작업
    for i, day in enumerate(day_lst):
        webtoon_day = pd.DataFrame()
        if i==0:
            day_toon = soup.find("div", attrs={"class":re.compile("^col")})
        else:
            day_toon = day_toon.find_next_sibling("div")

        cartoons = day_toon.find_all("a", attrs={"class":"title"})
        
        # 요일 별로 제목을 가져와 dataframe으로 만들기
        webtoon = []
        for cartoon in cartoons:
            title = cartoon.get_text()
            webtoon.append([title])
            webtoon_title = pd.DataFrame(webtoon)

        webtoon_rank = pd.concat([webtoon_rank, webtoon_title], axis=1, ignore_index=True)
    webtoon_rank.columns = [f'{day}요일' for day in day_lst]
    webtoon_rank.index = [f'{i+1}위' for i in webtoon_rank.index]
    
    # 원하는 요일만 볼 수 있음
    # 지정하지 않을 시, 전체 요일 출력
    if day == None:
      webtoon_rank = webtoon_rank
    else:
      webtoon_rank = webtoon_rank.loc[:, days]

    webtoon_rank.to_csv("webtoon_rank")

    return webtoon_rank
```
